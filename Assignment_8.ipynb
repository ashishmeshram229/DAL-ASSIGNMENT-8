```python
# Step 1: Import Required Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Optional: make plots look better
plt.style.use("seaborn-v0_8-muted")
plt.rcParams.update({'figure.dpi': 120})
```

```python
# Step 2: Load the Bike Sharing Dataset (hour.csv)

# Assuming the folder "bike+sharing+dataset" is in the same directory as this notebook
df = pd.read_csv("bike+sharing+dataset/hour.csv")

print("✅ Dataset loaded successfully!")
print("Shape of dataset:", df.shape)

# Display first few rows
display(df.head())
```

```python
# Step 3: Drop Irrelevant or Leakage Columns

drop_cols = ["instant", "dteday", "casual", "registered"]

df.drop(columns=drop_cols, inplace=True, errors="ignore")

print("Dropped columns:", drop_cols)
print("Shape after dropping columns:", df.shape)

# Quick check
display(df.head())
```

```python
# Step 4: Feature Engineering
# - Apply cyclical encoding for 'hr'
# - Apply One-Hot Encoding for 'season', 'weathersit', 'mnth'

# Cyclical encoding for 'hr' (hour of the day)
if "hr" in df.columns:
    df["hr_sin"] = np.sin(2 * np.pi * df["hr"] / 24)
    df["hr_cos"] = np.cos(2 * np.pi * df["hr"] / 24)
    df.drop(columns=["hr"], inplace=True)
    print("Applied cyclical encoding for 'hr' → hr_sin, hr_cos")

# One-Hot Encoding for selected categorical variables
categorical_features = ["season", "weathersit", "mnth"]
df = pd.get_dummies(df, columns=categorical_features, drop_first=False)

print("Applied One-Hot Encoding on:", categorical_features)
print("Shape after encoding:", df.shape)

# Preview encoded data
display(df.head())
```

```python
# Step 5: Inspect Dataset Shape, Missing Values, and Data Types

print("Dataset Overview After Preprocessing:\n")
print(df.info())

print("\nMissing Values per Column:")
print(df.isnull().sum())

print("\nFinal Dataset Shape:", df.shape)
```

```python
# Step 6: Visualize Target Variable 'cnt'

plt.figure(figsize=(8, 4))
plt.hist(df["cnt"], bins=40, color="#42a5f5", edgecolor="black", alpha=0.8)
plt.title("Distribution of Target Variable: cnt", fontsize=13, color="#263238")
plt.xlabel("Bike Rental Count (cnt)")
plt.ylabel("Frequency")
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Step 7: Summary

print("Data preprocessing and feature engineering completed successfully!")
print("Final number of features:", df.shape[1])
```

```python
#  Prepare X, y and choose split method
import numpy as np
from sklearn.model_selection import train_test_split, TimeSeriesSplit

# df is expected to be present in the notebook (output of Part A.1)
# If not, load the preprocessed file:
# df = pd.read_csv("bike+sharing+dataset/hour_preprocessed.csv")

target = "cnt"
if target not in df.columns:
    raise ValueError(f"Target column '{target}' not found in dataframe.")

X = df.drop(columns=[target]).copy()
y = df[target].copy()

print("X shape:", X.shape)
print("y shape:", y.shape)

# Choose split_type: 'time' for chronological split, 'random' for stratified/random split
split_type = "time"   # options: "time" or "random"
test_size = 0.20      # fraction for test set (used for random split); for time split it's the tail fraction
random_state = 42
```

```python
#  Perform the split (chronological by default)
if split_type == "time":
    n = len(X)
    test_n = int(np.ceil(n * test_size))
    train_n = n - test_n
    X_train = X.iloc[:train_n].reset_index(drop=True)
    X_test  = X.iloc[train_n:].reset_index(drop=True)
    y_train = y.iloc[:train_n].reset_index(drop=True)
    y_test  = y.iloc[train_n:].reset_index(drop=True)
    print(f"Chronological split: train={len(X_train)} rows, test={len(X_test)} rows")
else:
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=test_size, random_state=random_state, shuffle=True
    )
    print(f"Random split: train={len(X_train)} rows, test={len(X_test)} rows")

# Quick checks
print("\nTrain target statistics:")
print(y_train.describe().round(2))
print("\nTest target statistics:")
print(y_test.describe().round(2))
```

```python
# (Optional) Setup TimeSeriesSplit for cross-validation later
from sklearn.model_selection import TimeSeriesSplit
n_splits = 5
tscv = TimeSeriesSplit(n_splits=n_splits)
print(f"Created TimeSeriesSplit with {n_splits} splits (use tscv in model validation).")
```

```python
#Import required libraries for baseline models
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np

# Ensure X_train, X_test, y_train, y_test are already defined from Part A.2
print("Training samples:", X_train.shape[0])
print("Testing samples:", X_test.shape[0])
```

```python
# Train Decision Tree Regressor (max_depth = 6)
dt_model = DecisionTreeRegressor(max_depth=6, random_state=42)
dt_model.fit(X_train, y_train)

# Predictions
y_pred_dt = dt_model.predict(X_test)

# Evaluate using RMSE
rmse_dt = np.sqrt(mean_squared_error(y_test, y_pred_dt))
print(f"Decision Tree RMSE: {rmse_dt:.4f}")
```

```python
# Train Linear Regression Model
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)

# Predictions
y_pred_lr = lr_model.predict(X_test)

# Evaluate using RMSE
rmse_lr = np.sqrt(mean_squared_error(y_test, y_pred_lr))
print(f"Linear Regression RMSE: {rmse_lr:.4f}")
```

```python
# Compare models and set baseline
if rmse_dt < rmse_lr:
    baseline_model = "Decision Tree Regressor"
    baseline_rmse = rmse_dt
else:
    baseline_model = "Linear Regression"
    baseline_rmse = rmse_lr

print("\n✅ Baseline Model Selected:")
print(f"Model: {baseline_model}")
print(f"RMSE: {baseline_rmse:.4f}")
```

      Instead of averaging predictions like bagging or sequentially correcting errors like boosting,
      stacking learns <em>how to optimally blend the predictions</em> of diverse models to minimize overall error.

    </h3>
    <p style="margin: 0; color: #475569; font-size: 0.95em; line-height: 1.6;">
        Stacking creates a two-layer model structure:
        <br><strong>• Level-0 (Base Learners):</strong> Multiple independent models (e.g., KNN, Bagging, Gradient Boosting)
        each make predictions on the same data.<br>
        <strong>• Level-1 (Meta-Learner):</strong> A simple model (often Ridge or Linear Regression)
        takes these predictions as input features and learns how to combine them
        for the most accurate final output.
    </p>
  </div>

         How the Meta-Learner Works
    </h3>
    <p style="margin: 0; color: #475569; font-size: 0.95em; line-height: 1.6;">
        The meta-learner is trained on the <em>out-of-fold predictions</em> from each base model —
        meaning it never sees predictions from data the base model was trained on.
        This prevents data leakage and allows it to learn the best combination of base model outputs.
        <br><br>
        Essentially, the meta-learner acts like a “coach,” learning which model performs better
        for different situations and assigning appropriate weights to each one.
    </p>
  </div>

  <div style="
      background-color: #ffffff;
      border-radius: 10px;
      padding: 16px 18px;

  ">
    <h3 style="
        font-size: 1.05em;
        color: #0f172a;
        font-weight: 600;
        margin: 0 0 6px 0;
    ">
         Why Stacking Works
    </h3>
    <ul style="margin: 0; padding-left: 20px; color: #475569; font-size: 0.95em; line-height: 1.6;">
        <li>It leverages <strong>model diversity</strong> — different algorithms capture different patterns.</li>
        <li>Reduces both <strong>bias</strong> and <strong>variance</strong> by learning optimal blending weights.</li>
        <li>Produces <strong>more stable and accurate predictions</strong> than any single model alone.</li>
    </ul>
  </div>

  <p style="
      margin-top: 14px;
      color: #64748b;
      font-size: 0.9em;
  ">
      <em>In summary:</em> Stacking intelligently combines the “opinions” of multiple models through a meta-learner,
      resulting in a final model that generalizes better than individual base learners.
  </p>
</div>

```python
# imports
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Ridge
from sklearn.ensemble import StackingRegressor, BaggingRegressor, GradientBoostingRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({"figure.dpi":120})
```

```python
#  Define Level-0 base learners (with a scaler for KNN)
knn_pipe = Pipeline([
    ("scaler", StandardScaler()),
    ("knn", KNeighborsRegressor(n_neighbors=8, weights="distance", n_jobs=-1))
])

bag_reg = BaggingRegressor(
    estimator=DecisionTreeRegressor(max_depth=6, random_state=42),
    n_estimators=50,
    random_state=42,
    n_jobs=-1
)

gbr_reg = GradientBoostingRegressor(
    n_estimators=300,
    learning_rate=0.05,
    max_depth=3,
    subsample=0.9,
    random_state=42
)

estimators = [
    ("knn", knn_pipe),
    ("bagging", bag_reg),
    ("gboost", gbr_reg)
]

meta_learner = Ridge(alpha=1.0, random_state=42)

print("Level-0 learners and Level-1 meta-learner defined.")
for name, est in estimators:
    print(f" - {name}: {est.__class__.__name__}")
print("Meta-learner:", meta_learner.__class__.__name__)
```

```python
#  Build and train the StackingRegressor
stack = StackingRegressor(
    estimators=estimators,
    final_estimator=meta_learner,
    cv=5,            # OOF predictions for training meta-learner
    n_jobs=-1,
    passthrough=False  # if True, original features are also passed to meta-learner
)

stack.fit(X_train, y_train)
print("✅ StackingRegressor trained.")
```

```python
# Evaluate Stacking on test set (RMSE) and compare with previous results (if any)
y_pred_stack = stack.predict(X_test)
rmse_stack = np.sqrt(mean_squared_error(y_test, y_pred_stack))
print(f"Stacking Regressor RMSE: {rmse_stack:.4f}")

# If baseline rmse variables exist, show comparisons
for name, var in [("Decision Tree", "rmse_dt"), ("Bagging", "rmse_bag"), ("Gradient Boosting", "rmse_gbr")]:
    if var in globals():
        print(f"{name} RMSE: {globals()[var]:.4f}    Δ vs Stacking: {globals()[var] - rmse_stack:.4f}")
```
